# üìç **Quais s√£o os pa√≠ses com mais casos de COVID-19?**
Vamos analisar os dados exclusivos da vari√°vel `location`, que cont√©m a localiza√ß√£o do registro, a fim de identificar esses pa√≠ses.

```python
# verificar os dados √∫nicos da vari√°vel
df.location.unique()
```

Podemos observar que nem todos os registros se referem apenas a pa√≠ses espec√≠ficos; existem tamb√©m dados como '**World**', que representa o mundo inteiro, e '**High income**', que se refere a pessoas de alta renda. Al√©m disso, h√° registros para continentes e outros valores. Esses dados que n√£o s√£o relevantes para nossa an√°lise est√£o associados ao campo `iso_code`, que cont√©m a palavra "**OWID**". Essa rela√ß√£o nos permite agrupar esses tipos de dados.

```python
# verificar as cinco primeiras tuplas cujo iso_code cont√©m 'OWID'
df.loc[df.iso_code.str.contains("OWID")].head(5)
```

Para dar **continuidade** √† **an√°lise** em foco e obter os *cinco pa√≠ses com maior n√∫mero de casos*, vamos remover as entradas que correspondem a esse agrupamento espec√≠fico.

```python
# c√≥pia do dataset
df_covid = df.copy()
# apagar tuplas nas quais o iso_code contenha "OWID"
df_covid.drop(df_covid.index[df_covid.iso_code.str.contains("OWID")].values, inplace = True)
```

Ap√≥s a devida prepara√ß√£o dos dados, estaremos prontos para apresentar o resultado dos pa√≠ses com os maiores casos de **COVID-19**.

```python
df_covid.groupby(['location']).total_cases.sum().sort_values(ascending = False)[:5]
```

A seguir, examinaremos a propor√ß√£o de casos em rela√ß√£o √† quantidade total de casos registrados nesses pa√≠ses.

```python
(df_covid.groupby(['location']).total_cases.sum()/df_covid.total_cases.sum() * 100).sort_values(ascending = False)[:5]
```

Plotaremos um gr√°fico de barras para a melhor visualiza√ß√£o do resultado.

```python
# Plotando gr√°ficos para pa√≠ses com mais casos registrados at√© a data atual
fig, ax = plt.subplots()

# Obtendo os valores e √≠ndices dos pa√≠ses com mais casos
x = (df_covid.groupby(['location']).total_cases.sum().sort_values(ascending = False)[:5]).values
y = (df_covid.groupby(['location']).total_cases.sum().sort_values(ascending = False)[:5]).index

# Gr√°fico da quantidade de casos de COVID-19 por pa√≠s
ax = sns.barplot(x = x, y = y, ax = ax, palette = "magma")
ax.grid(False)
ax.set_ylabel('Pa√≠ses')
ax.set_xlabel('Casos de COVID-19')
ax.set_title("Top 5 Pa√≠ses com Mais Casos de COVID-19");

plt.tight_layout()
plt.show()
```

```python
# Calcular a soma dos casos dos cinco primeiros pa√≠ses
total_cases_top5 = df_covid.groupby(['location']).total_cases.sum().sort_values(ascending=False)[:5].sum()

# Calcular a soma de todos os casos
total_cases_all = df_covid['total_cases'].sum()

# Calcular a porcentagem
percentage = (total_cases_top5 / total_cases_all) * 100

print(f"A porcentagem que os cinco pa√≠ses com mais casos representam em rela√ß√£o ao restante √©: {percentage:.2f}%")
```

# üìç H√° Correla√ß√£o entre Popula√ß√£o Idosa e Mortes por COVID-19?

Compreender a rela√ß√£o entre a parcela da popula√ß√£o com 65 anos ou mais e o n√∫mero total de mortes por COVID-19 √© fundamental para entender os fatores de risco associados √† gravidade da doen√ßa. Embora a maioria dos casos apresente sintomas leves a moderados, certos grupos enfrentam um risco maior de manifesta√ß√µes cl√≠nicas graves, incluindo idosos, indiv√≠duos do sexo masculino e pessoas com comorbidades.

A influ√™ncia da idade √© especialmente significativa, sendo um dos principais fatores de risco para casos graves e desfechos negativos da COVID-19. Para avaliar essa rela√ß√£o, utilizamos o coeficiente de correla√ß√£o, uma medida estat√≠stica que quantifica a associa√ß√£o entre duas vari√°veis.

"O coeficiente de correla√ß√£o √© uma m√©trica estat√≠stica que avalia a rela√ß√£o entre duas vari√°veis, revelando o grau de associa√ß√£o √† medida que o valor de uma vari√°vel se modifica."

* Iremos plotar uma **matriz de correla√ß√£o** para observar a medida obtida entre as vari√°veis `aged_65_older` e `total_deaths`.
* Como o total de mortes √© um valor absoluto, enquanto a quantidade de pessoas acima de 65 anos √© uma parcela da popula√ß√£o, para correlacion√°-los iremos transformar a vari√°vel `total_deaths`, dividindo-a pelo n√∫mero da popula√ß√£o, criando a vari√°vel `percentual_mortes`.
* Vamos relacionar o percentual de mortes (em rela√ß√£o √† popula√ß√£o total) com o percentual de pessoas com 65 anos ou mais.
* Depois, faremos uma an√°lise semelhante para as vari√°veis `gdp_per_capita` e `total_deaths_per_million`. Seguiremos os seguintes passos:

Os passos a seguir guiam nossa an√°lise:

1. Cria√ß√£o da matriz de correla√ß√£o.
2. Exibi√ß√£o dos resultados obtidos.
3. Realiza√ß√£o de um teste de hip√≥teses.
4. Plotagem de um gr√°fico de correla√ß√£o.
5. Plotagem de um gr√°fico de dispers√£o.

Ao final, nossa an√°lise proporcionar√° uma vis√£o mais clara sobre a associa√ß√£o entre a parcela idosa da popula√ß√£o e o n√∫mero de mortes por COVID-19.

```python
# fazer uma c√≥pia do dataframe
df_corr = df_covid
df_corr.dropna(subset=['total_deaths'], inplace = True)

# criar vari√°vel que √© o percentual de mortes por poupula√ß√£o
div_column = df_corr['total_deaths']/df_corr['population']
df_corr['percentual_mortes'] = div_column

# criar uma matriz de correla√ß√£o
corr_df = df_corr[['aged_65_older','percentual_mortes', 'gdp_per_capita', 'total_deaths_per_million']].corr(method='pearson');
# mostrar a matriz de correla√ß√£o
display(corr_df);
```

```python
# plotar o gr√°fico de correla√ß√£o
sns.heatmap(corr_df, fmt='.2f',square = True, annot = True);
```

Vamos conduzir um teste de hip√≥teses com um n√≠vel de signific√¢ncia de 5% para determinar se h√° uma correla√ß√£o entre `aged_65_older` e `percentual_mortes`.

Teste de Hip√≥teses ‚Äî `aged_65_older` e `percentual_mortes`

œÅ = coeficiente de correla√ß√£o linear de Pearson populacional

**Hip√≥tese Nula H0** : œÅ = 0

A hip√≥tese nula afirma que n√£o existe uma rela√ß√£o linear significativa entre a parcela da popula√ß√£o com 65 anos ou mais e o percentual de mortes. Em outras palavras, o coeficiente de correla√ß√£o linear entre essas vari√°veis √© zero.

**Hip√≥tese Alternativa HA **: œÅ ‚â† 0

A hip√≥tese alternativa sugere que existe um coeficiente de correla√ß√£o linear significativo entre as duas vari√°veis.

```python
# vari√°veis necess√°rias para o c√°lculo
r = 0.473442
r_2 = r ** 2
n = df_covid.shape[0]
print("R = ",r)
print("R¬≤ = ",round(r_2 * 100,2), "%")
print("N = ", n)
```

A vari√°vel `aged_65_older` afeta o percentual de mortes em **22.41%**.

```python
# T-Calculado
tc = (r * math.sqrt(n-2))/math.sqrt(1 - r_2)
print("Tc: ", round(tc,4))
```

```python
# graus de liberdade n-2 com 5% de signfic√¢ncia.
t = stats.t.ppf(1-0.025, n - 2)
print ("Valor na t de student: ", round(t,2))
```

Uma vez que **tc > t**, rejeitamos a hip√≥tese nula e conclu√≠mos que, com um n√≠vel de signific√¢ncia de **5%**, h√° uma correla√ß√£o positiva entre o n√∫mero total de pessoas com mais de 65 anos e o percentual de mortes.

A **pr√≥xima etapa** envolve a cria√ß√£o de um gr√°fico de dispers√£o que relaciona essas duas vari√°veis a partir do m√™s de agosto.

```python
# Filtrando os dados a partir da √∫ltima data dispon√≠vel
df_filtrado = df_covid.loc[df_covid.date >= '2023-08-16']

# Criando a figura do gr√°fico e ajustando o tamanho
fig = plt.gcf()
fig.set_size_inches(15, 8)

# Plotando um gr√°fico de dispers√£o com paleta colorida
ax = sns.scatterplot(data=df_filtrado, x="aged_65_older", y="percentual_mortes", hue="continent", palette="pastel", s=100)

# Definindo os r√≥tulos dos eixos
ax.set(xlabel='Parcela da Popula√ß√£o com mais de 65 anos', ylabel='Percentual de Mortes')

# Definindo o t√≠tulo do gr√°fico
ax.set_title(f"Rela√ß√£o entre Parcela da Popula√ß√£o com mais de 65 anos e Percentual de Mortes ‚Äî {last_date.month}/{last_date.year}")

# Desativando as grades no gr√°fico
ax.grid(False)

# Exibindo a legenda
ax.legend(title='Continente')

# Exibindo o gr√°fico
plt.show()
```

Observa-se de maneira geral que √† medida que a propor√ß√£o da popula√ß√£o com idade acima de 65 anos aumenta, o percentual total de mortes tamb√©m tende a aumentar.

Agora, investigaremos a poss√≠vel correla√ß√£o entre a vari√°vel `gdp_per_capita` (PIB per capita) e `total_deaths_per_million` (n√∫mero total de mortes por milh√£o de habitantes). Para isso, realizaremos um teste de hip√≥teses com um n√≠vel de signific√¢ncia de 5% e plotaremos um gr√°fico de dispers√£o. A vari√°vel `gdp_per_capita` representa a economia de um pa√≠s, permitindo-nos explorar a rela√ß√£o entre a sa√∫de econ√¥mica e o n√∫mero de mortes.

**Teste de Hip√≥teses ‚Äî gdp_per_capita e total_deaths_per_million**
œÅ = coeficiente de correla√ß√£o linear de Pearson populacional

- **Hip√≥tese Nula H0 : œÅ = 0**
A hip√≥tese nula assume que n√£o existe rela√ß√£o linear entre o PIB per capita e o n√∫mero total de mortes. Ou seja, o coeficiente de correla√ß√£o linear entre essas duas vari√°veis √© nulo.

- **Hip√≥tese Alternativa HA : œÅ ‚â† 0**
A hip√≥tese alternativa sugere que existe uma correla√ß√£o linear significativa entre o PIB per capita e o n√∫mero total de mortes.

Vamos prosseguir com a an√°lise plotando um gr√°fico de dispers√£o para visualizar melhor essa poss√≠vel rela√ß√£o.

```python
# vari√°veis necess√°rias para o c√°lculo
r = 0.183923
r_2 = r ** 2
n = df_covid.shape[0]
print("R = ",r)
print("R¬≤ = ",round(r_2 * 100,2), "%")
print("N = ", n)
```

Pode-se observar que a vari√°vel` gdp_per_capita` afeta o n√∫mero total de mortes em 3.38%



```python
# T-Calculado
tc = (r * math.sqrt(n-2))/math.sqrt(1 - r_2)
print("Tc: ", round(tc,4))
```

```python
# graus de liberdade n-2 com 5% de signfic√¢ncia.
t = stats.t.ppf(1-0.025, n - 2)
print ("Valor na t de student: ", round(t,2))
```

Dado que o valor de tc √© maior que o valor cr√≠tico t, rejeitamos a hip√≥tese nula. Conclu√≠mos, portanto, que com um n√≠vel de signific√¢ncia de 5%, h√° uma correla√ß√£o positiva entre o PIB do pa√≠s e o n√∫mero total de mortes. Em outras palavras, h√° ind√≠cios estat√≠sticos de que o crescimento do PIB per capita est√° associado a um aumento no n√∫mero total de mortes por milh√£o de habitantes.

Essa correla√ß√£o aponta para a possibilidade de que as vacinas possam ter contribu√≠do para a redu√ß√£o da taxa de mortalidade.

A fim de investigar uma poss√≠vel correla√ß√£o entre essas vari√°veis, ser√° constru√≠da uma matriz de correla√ß√£o entre o percentual de mortes em rela√ß√£o ao n√∫mero de casos e o percentual de vacinados em rela√ß√£o √† popula√ß√£o.

```python
# fazer uma c√≥pia do dataframe
df_corr = df_brasil

# criar vari√°vel que √© o n√∫mero de mortes por n√∫mero de casos
div_column = df_corr['total_deaths']/df_corr['total_cases']
df_corr['percentual_mortes'] = div_column

# criar vari√°vel que √© o n√∫mero de pessoas vacinadas pela popula√ß√£o
div_column = df_corr['people_vaccinated']/df_corr['population']
df_corr['percentual_vacinados'] = div_column

# criar uma matriz de correla√ß√£o
corr_df = df_corr[['percentual_mortes', 'percentual_vacinados']].corr(method='pearson');

# mostrar a matriz de correla√ß√£o
display(corr_df);
```
